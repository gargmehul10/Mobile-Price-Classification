```python
#creating a map for values of RAM, as it is highly correlated to the price range
nps=d1['ram'].tolist()
for x in nps:
    if x<1000:
        x=0
    elif x<2000:
        x=1
    elif x<3000:
        x=2
    else:
        x=3
d1['ram']=nps
```

```python
from sklearn.model_selection import train_test_split #for creating training and testing data out of the entire training dataset
from sklearn import metrics #for measuring accuracy
#importing classifier from scikit-learn library
from sklearn.neighbors import KNeighborsClassifier
```

```python
train,test=train_test_split(d1,test_size=0.2)
print(train.shape)
print(test.shape)
```

```python
d2=pd.read_csv('test.csv')
```

```python
d2.info()
```

```python
d2.drop('id', axis='columns', inplace=True)
#ID causes inaccuracy of results
```

```python
#creating a map for values of RAM, as it is highly correlated to the price range
nps=d1['ram'].tolist()
for x in nps:
    if x<1000:
        x=0
    elif x<2000:
        x=1
    elif x<3000:
        x=2
    else:
        x=3
d1['ram']=nps
```

```python
#removing some internally correlated features(or columns) from the dataset
train_X=train[['battery_power','blue','clock_speed','dual_sim','int_memory','m_dep','mobile_wt','n_cores','pc','px_height','px_width','ram','sc_w','talk_time','three_g','touch_screen','wifi']]
train_Y=train.price_range
test_X=test[['battery_power','blue','clock_speed','dual_sim','int_memory','m_dep','mobile_wt','n_cores','pc','px_height','px_width','ram','sc_w','talk_time','three_g','touch_screen','wifi']]
test_Y=test.price_range
```

```python
d2=d2[['battery_power','blue','clock_speed','dual_sim','int_memory','m_dep','mobile_wt','n_cores','pc','px_height','px_width','ram','sc_w','talk_time','three_g','touch_screen','wifi']]
```

```python
model=KNeighborsClassifier(n_neighbors=50, weights='distance')
model.fit(train_X,train_Y) 
prediction=model.predict(test_X) 
print(metrics.accuracy_score(prediction,test_Y))
```

```python
pred=model.predict(d2)
out=pd.DataFrame({'price_range':pred.tolist()})
out.index=pd.MultiIndex.from_arrays([np.arange(1,1001)],names=['id'])
out.to_csv('out.csv')
```
